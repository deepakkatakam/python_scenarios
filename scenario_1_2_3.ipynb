#Standardizing contact details in an Excel file for a marketing campaign using  customer data

```python
countrycodes = pd.read_csv(r"C:\Users\Deepak\Desktop\python_practice\Country-codes.csv", encoding='ISO-8859-1')

customerdata['country_code'] = customerdata['address'].str.slice(-8, -6)
customerdata['country_code']
customerdata['phone'] = customerdata['phone'].astype(str)
```

```python
for i in range(len(customerdata)):
    code = customerdata.loc[i, 'country_code']
    for j in range(len(countrycodes)):
        if code == countrycodes.loc[j, 'Country_code']:
            dial = countrycodes.loc[j, 'International_dialing']
            customerdata.loc[i, 'phone'] = (str(dial) + " "+ customerdata.loc[i, 'phone'])
customerdata
```
